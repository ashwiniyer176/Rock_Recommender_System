```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df[:][2:]
```

```python
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df[df.columns[2:]])
scaled_df=pd.DataFrame(scaled_df,columns=df.columns[2:])
```

```python
scaled_df
```

```python
scaled_df['name']=df['name']
scaled_df['artist']=df['artist']
```

```python
scaled_df
```

```python
KM=KMeans(n_clusters=4)
```

```python
KM.fit_predict(scaled_df.drop(columns=["name","artist"]))
```

```python
sse=[]
for i in range(2,15):
    KM=KMeans(n_clusters=i)
    KM.fit_predict(scaled_df.drop(columns=["name","artist"]))
    sse.append(KM.inertia_)
```

```python
sse
```

```python
def plotElbowGraph(sseValues):
    plt.figure(figsize=(10,8))
    plt.plot(range(2,len(sseValues)+2),sseValues)
    plt.xticks(range(2,len(sseValues)+2))
    plt.title("Elbow Method")
    plt.xlabel("No of Clusters")
    plt.ylabel('SSE Values')
```

```python
plotElbowGraph(sse)
```

```python
def formatName(songName):
    songName=songName.split('-')
    print(songName[0].lower())
    return songName[0].lower()
```

```python
scaled_df.name=scaled_df['name'].apply(formatName)
scaled_df.artist=scaled_df.artist.apply(formatName)
```

```python
scaled_df
```

```python
def findSongsSimilarTo(songName,numberOfSongs,data=scaled_df):
    label=data.loc[data.name==songName.lower()].values.tolist()[0][-1]
    print(label)
    songs=data.loc[data.Song_Label==label].values.tolist()
    count=1
    for song in songs:
        if(count<=numberOfSongs):
            print(song[-3:-1])
            count+=1
```

```python
findSongsSimilarTo("Back in Black",5)
```
